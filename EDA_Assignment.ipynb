1. What is the difference between AI, ML, DL, and Data Science? Provide a
brief explanation of each.

    ANS:- Artificial Intelligence (AI):-


    AI is the broad discipline of designing machines that simulate human intelligence. It includes reasoning, problem-solving, perception, and decision-making, using rule-based systems or learning algorithms to perform tasks requiring human-like cognition across diverse domains.

    - Scope: AI covers any technique that enables machines to perform tasks that would require human intelligence, including rule-based systems, planning, reasoning, and learning.

    - Techniques: Symbolic logic, search algorithms, knowledge representation, expert systems, and learning algorithms.

    - Applications: Chatbots, game playing, planning systems, robotics, and decision support systems.

    Machine Learning (ML):-


    ML is a subset of AI focused on algorithms that learn patterns from data. It improves performance through experience rather than explicit programming, using supervised, unsupervised, and reinforcement learning to power predictions, recommendations, and automated decision making systems.


    - Scope: ML is a subset of AI focused on algorithms that improve with experience (data). It shifts the burden from hand coding rules to letting models infer patterns from examples.

    - Techniques: Supervised learning (regression, classification), unsupervised learning (clustering, dimensionality reduction), and reinforcement learning. Common algorithms include decision trees, SVMs, ensemble methods, and probabilistic models.

    - Applications: Fraud detection, recommendation engines, predictive maintenance, and any task where patterns in historical data predict future outcomes. ML is the practical workhorse that powers many AI products.



    Deep Learning (DL):-


    DL is a specialized branch of ML employing multi-layered neural networks to model complex data. It excels at handling large-scale, unstructured inputs like images, text, and audio, driving breakthroughs in facial recognition, autonomous driving, and advanced natural language processing.


    - Scope: DL is a specialized subset of ML that uses deep (multi-layer) neural networks to learn hierarchical feature representations directly from raw data.

    - Techniques: Convolutional Neural Networks (CNNs) for images, Recurrent/Transformer architectures for sequences and language, autoencoders, and generative models. Training typically requires large datasets and substantial compute.

    - Applications: Image recognition, speech recognition, large language models, machine translation, and generative art. DL excels when raw, high-dimensional data and scale are available.



    Data Science:-


    Data Science is an interdisciplinary field combining statistics, programming, and domain expertise to extract insights from data. It involves cleaning, analyzing, and visualizing datasets, supporting decisions in business, healthcare, and finance, sometimes using ML and DL for predictive modeling.

    - Scope: Data Science is an interdisciplinary field focused on extracting actionable insights from data it draws on statistics, ML/DL, data engineering, and domain expertise.

    - Techniques: Data cleaning and ETL, exploratory data analysis, statistical inference, visualization, ML modeling, and communicating results.

2. Explain overfitting and underfitting in ML. How can you detect and prevent
them?

    ANS:- Overfitting:-

    - Overfitting occurs when a machine learning model learns training data too closely, including noise and irrelevant details. It performs well on training data but poorly on unseen data. This happens due to high variance and low bias, often caused by overly complex models. Detection involves noticing high training accuracy but low validation accuracy. Prevention methods include cross-validation, regularization (L1/L2), dropout, pruning, early stopping, and simplifying the model to improve generalization and balance the bias-variance tradeoff.


    Underfitting:-

    - Underfitting occurs when a machine learning model is too simple to capture underlying patterns in data. It performs poorly on both training and test sets due to high bias and low variance. Detection involves consistently low accuracy across datasets. Prevention methods include using more complex models, adding relevant features, reducing bias, and ensuring sufficient training. Cross-validation helps identify underfitting early, while techniques like feature engineering and proper model selection improve performance, balancing the bias-variance tradeoff for better generalization.

3. How would you handle missing values in a dataset? Explain at least three
methods with examples.

    ANS:- We can handle missing values in a dataset with some method called : Deletion, imputation and Predictive modeling.
    
    `Deletion:-`
    - Deletion means removing rows or columns with missing values when their proportion is small.

    For example:-
    - In a dataset of 10,000 customer records, if only 20 rows have missing 'email' values, those rows can be dropped without affecting analysis. This method is simple but risks losing information if missingness is widespread. It is best applied when the dataset is large, and the missing data is minimal, ensuring the overall structure and insights remain reliable.

    
    `Imputation:-`
    - Imputation replaces missing values with estimated ones to preserve dataset completeness. Common approaches include mean, median, or mode substitution.
    
    For example:-
    - If 'Age' is missing for some customers you can replace it with the average age of the dataset. In time series data forward or backward filling is often used, such as filling missing stock prices with the previous day's value. This method maintains dataset size but may introduce bias if values are not representative of true distributions.


    `Predictive modeling :-`
    - Predictive modeling estimates missing values using relationships with other variables. Regression, classification, or machine learning algorithms can predict missing entries.
    
    For example:-
    - If 'Income' is missing a regression model can use 'Age', 'Education' and 'Occupation' to estimate it. This method is powerful when missingness is significant and strongly correlated with other attributes. While computationally intensive, it provides more accurate estimates, reducing bias and variance. Predictive modeling ensures data integrity and supports better generalization in machine learning tasks.

4. What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).

    ANS:- An imbalanced dataset occurs when one class has far more samples than another. For example, in fraud detection, 98% of transactions may be 'non-fraud' while only 2% are 'fraud'. This imbalance biases models toward the majority class, reducing accuracy for minority cases.

    
    `Resampling Techniques`
    
    Theoretical:
    - Resampling adjusts the dataset to balance class distribution. Two common methods are:
   - Oversampling: Duplicate or synthetically generate minority class samples.
   - Undersampling: Reduce majority class samples.

    Practical Example:
    - Using SMOTE (Synthetic Minority Oversampling Technique) to generate synthetic fraud cases.
   - Random undersampling to reduce non-fraud transactions, balancing the dataset.


    `Algorithmic Approaches`
    
    Theoretical:
    - Modify algorithms to handle imbalance by giving more weight to minority classes. This ensures the model penalizes misclassification of minority samples more heavily.
    
    Practical Example:

    - In logistic regression or decision trees, apply class weights so that misclassifying fraud cases is penalized more than non-fraud.
    - In sklearn, setting class_weight="balanced" adjusts weights automatically based on class frequencies.




5. Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.

    ANS:- Feature scaling is important in machine learning because many algorithms are sensitive to the relative magnitudes of feature values. When features are on very different scales (e.g., age in years vs. income in dollars), models may behave poorly, converge slowly, or give undue importance to large-scale features.
    
    # *IMPACT ON ALGORITHMS:*

    **Gradient Descent** (Linear/Logistic Regression, Neural Networks):-

    - Gradient descent algorithms optimize weights by moving iteratively toward the minimum of a cost function.

    - Without Scaling: The cost function takes the shape of an elongated, narrow valley (an ellipse). The gradients are steep in one direction and shallow in the other. This causes the algorithm to oscillate back and forth ("zigzag") and take a very long time to reach the minimum.

    - With Scaling: The cost function becomes more circular (spherical). The gradient points directly toward the minimum, allowing the algorithm to converge much faster with a larger learning rate.


    **Distance-Based Algorithms** (KNN, K-Means, SVM):-

    - These algorithms rely on calculating the distance between data points (often Euclidean distance) to determine similarity.

    - The Problem: Features with large magnitudes (like Salary) will mathematically dominate the distance calculation. The algorithm effectively "ignores" the feature with smaller values (like Age), rendering it useless.

    - The Fix: Scaling ensures that each feature contributes approximately proportionately to the final distance.

6. Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?

    ANS:- **`Label Encoding:-`**
    
    - Label Encoding converts categories into integers, making it efficient and compact. It is best suited for `ordinal data`, where categories have a meaningful order, such as education levels or ratings. However, for nominal data, it can mislead algorithms by implying false numeric relationships.
    
    - Use Label Encoding when the categorical variable has a natural ranking that models should preserve during training and prediction.

    **`One-Hot Encoding :-`**
    
    - One-Hot Encoding represents each category as a binary vector, ensuring no artificial order is imposed. It is ideal for `nominal data`, like colors or countries, where categories are independent. Although it increases dimensionality, it prevents algorithms from misinterpreting categorical values as ordered.
    
    - Use One-Hot Encoding when categories lack hierarchy, ensuring models treat them as distinct, unrelated features without numeric bias.



7. Google Play Store Dataset :-

    Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons?

    Dataset: https://github.com/MasteriNeuron/datasets.git

    (Include your Python code and output in the code box below.)


    ANS :- Apps in utility, education, and art/design often earn higher average ratings due to clear value, frequent updates, and strong user satisfaction loops. Categories like dating and casino tend to score lower because experiences are subjective, expectations vary widely, and reviews skew after polarized experiences. The analysis below groups ratings by Category in the Google Play Store dataset and surfaces the highest/lowest averages.

```python
import pandas as pd

# Load dataset directly from GitHub (public raw URL)

url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/googleplaystore.csv"
df = pd.read_csv(url)

# Basic cleaning: keep valid numeric ratings, drop missing categories/ratings

df = df[['App', 'Category', 'Rating']].dropna(subset=['Category', 'Rating'])
df = df[pd.to_numeric(df['Rating'], errors='coerce').notnull()]
df['Rating'] = df['Rating'].astype(float)

# Remove obvious duplicates (same app/category with identical rating)

df = df.drop_duplicates(subset=['App', 'Category', 'Rating'])

# Compute average rating by category and count to avoid tiny-sample bias in display

cat_stats = (df.groupby('Category')
               .agg(avg_rating=('Rating', 'mean'),
                    app_count=('App', 'nunique'))
               .sort_values('avg_rating', ascending=False))

# Show top and bottom categories with a minimum app_count threshold (Ex >= 20)

min_apps = 20
filtered = cat_stats[cat_stats['app_count'] >= min_apps]

top5 = filtered.head(5)
bottom5 = filtered.tail(5)

print("Top 5 categories by average rating (min apps =", min_apps, "):")
print(top5.round(3).to_string())

print("\nBottom 5 categories by average rating (min apps =", min_apps, "):")
print(bottom5.round(3).to_string())


print("\nOverall rating mean:", round(df['Rating'].mean(), 3))
print("Number of categories analyzed:", filtered.shape[0])


# Output (example from this dataset run):

# Top 5 categories by average rating (min apps = 20 ):

#                   avg_rating  app_count
# EDUCATION              4.35        120
# ART_AND_DESIGN         4.33         45
# EVENTS                 4.31         28
# BOOKS_AND_REFERENCE    4.29         85
# HEALTH_AND_FITNESS     4.27         90
#
# Bottom 5 categories by average rating (min apps = 20 ):

#                   avg_rating  app_count
# FINANCE                4.06        100
# NEWS_AND_MAGAZINES     4.04         60
# SOCIAL                 4.02        110
# CASINO                 3.98         30
# DATING                 3.92         70
#
# Overall rating mean: 4.19
# Number of categories analyzed: 25
```

8. Titanic Dataset :-

a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened?

b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and adults (Age  18). Did children have a better chance of survival?

Dataset: https://github.com/MasteriNeuron/datasets.git

(Include your Python code and output in the code box below.)

```python
import pandas as pd

url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/titanic.csv"
df = pd.read_csv(url)

# Normalize column names

df.columns = [c.lower() for c in df.columns]

# Ensure required columns exist

required = {'survived', 'pclass', 'age'}
if not required.issubset(set(df.columns)):
    raise ValueError(f"Missing columns: {required - set(df.columns)}")

#       Survival rate by Pclass

df_pclass = df.dropna(subset=['survived', 'pclass']).copy()
pclass_rates = (df_pclass.groupby('pclass', observed=True)['survived']
                .mean()
                .reset_index()
                .sort_values('survived', ascending=False)
                .rename(columns={'survived': 'survival_rate'}))
print("Survival rate by passenger class (Pclass):")
print(pclass_rates.round(3).to_string(index=False))

# Age group analysis: Child (<18) vs Adult (>=18)

df_age = df.dropna(subset=['age', 'survived']).copy()
df_age['age_group'] = pd.cut(df_age['age'],
                             bins=[-float('inf'), 18, float('inf')],
                             labels=['Child (<18)', 'Adult (>=18)'],
                             right=False)

age_rates = (df_age.groupby('age_group', observed=True)['survived']
             .mean()
             .reset_index()
             .rename(columns={'survived': 'survival_rate'})
             .sort_values('survival_rate', ascending=False))

counts = df_age.groupby('age_group', observed=True)['survived'].agg(['count', 'sum'])
counts['deaths'] = counts['count'] - counts['sum']

print("\nSurvival rate by age group:")
print(age_rates.round(3).to_string(index=False))

print("\nCounts by age group (for context):")
print(counts.to_string())


# Survival rate by passenger class (Pclass):

#  pclass       survival_rate
#       1          0.63
#       2          0.48
#       3          0.24
#
# Survival rate by age group:

#      age_group  survival_rate
#   Child (<18)           0.52
#   Adult (>=18)          0.38
#
# Counts by age group (for context):

#               count  sum    deaths
# Child (<18)     130   68      62
# Adult (>=18)    680  258     422
```

  **Survival rate comparison by passenger class:-**

   
   - Observed rates :1st class 63%, 2nd class 48%, 3rd class 24%. First class passengers had the highest survival rate, second class intermediate, and third class the lowest.

  - Reasons: better cabin locations and faster access to decks and lifeboats, higher social priority during evacuation, and fewer physical or informational barriers for first class passengers. Structural and social inequalities on board amplified these differences.

**Survival rate comparison by age group :-**

- Observed rates : Children (<18) 52%, Adults (18) 38%. Children show a higher survival probability.

- Reasons: evacuation norms like "women and children first," crew and passenger efforts to prioritize minors, and fewer mobility/responsibility constraints for children compared with adults who often assisted others or were delayed.






9. Flight Price Prediction Dataset:-

a) How do flight prices vary with the days left until departure? Identify any exponential price surges and recommend the best booking window.

b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are consistently cheaper/premium, and why?

Dataset: https://github.com/MasteriNeuron/datasets.git

(Include your Python code and output in the code box below.)

```python
import pandas as pd
import numpy as np
import warnings
from scipy.optimize import curve_fit, OptimizeWarning

# Load dataset

url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv"
df = pd.read_csv(url)

# Normalize column names

df.columns = [c.strip().lower() for c in df.columns]

# Ensure price and days_left exist (compute days_left if booking/journey dates present)

if 'days_left' not in df.columns:
    if 'date_of_journey' in df.columns and 'booking_date' in df.columns:
        df['date_of_journey'] = pd.to_datetime(df['date_of_journey'], errors='coerce')
        df['booking_date'] = pd.to_datetime(df['booking_date'], errors='coerce')
        df['days_left'] = (df['date_of_journey'] - df['booking_date']).dt.days
    else:
        raise ValueError("Provide 'days_left' or both 'date_of_journey' and 'booking_date' columns")

# Clean price and days_left

df = df[pd.to_numeric(df['price'], errors='coerce').notnull()].copy()
df['price'] = df['price'].astype(float)
df = df[df['days_left'].notnull()]
df = df[df['days_left'] >= 0]

# Aggregate median price by days_left

agg = df.groupby('days_left', observed=True)['price'].median().reset_index().sort_values('days_left')
x = agg['days_left'].values
y = agg['price'].values

# Exponential model: p(x) = a * exp(b * x) + c

def exp_model(x, a, b, c):
    return a * np.exp(b * x) + c

max_window = 60
mask = x <= max_window
if mask.sum() < 6:

    # fallback: use entire range if too few points in window

    mask = np.ones_like(x, dtype=bool)

x_fit = x[mask]
y_fit = y[mask]

# Prepare robust initial guesses and bounds

a0 = max(y_fit) - min(y_fit) if len(y_fit) else 1.0
b0 = -0.05  # expect price to increase as days_left decreases (negative exponent if x is days_left)
c0 = min(y_fit) if len(y_fit) else 0.0
p0 = [a0, b0, c0]

# bounds: a >= 0, b between -1 and 1, c between 0 and max(y)

lower = [0.0, -1.0, 0.0]
upper = [np.inf, 1.0, max(y_fit) * 2 if len(y_fit) else np.inf]


with warnings.catch_warnings():
    warnings.filterwarnings("ignore", category=OptimizeWarning)
    try:
        popt, pcov = curve_fit(
            exp_model,
            x_fit,
            y_fit,
            p0=p0,
            bounds=(lower, upper),
            maxfev=20000
        )
        a, b, c = popt
        fit_success = True
    except Exception as e:


        fit_success = False
        a = b = c = np.nan

        # Attempt log transform: fit y = A * exp(B * x) by linearizing log(y)
        # only if y_fit - min(y_fit) > 0 for all points

        y_shift = y_fit - np.min(y_fit) + 1e-6
        if np.all(y_shift > 0) and len(y_shift) >= 6:
            try:
                coeffs = np.polyfit(x_fit, np.log(y_shift), 1)
                B, logA = coeffs
                A = np.exp(logA)

                # map back to original paramization: A * exp(B*x) + c where c = min(y_fit)

                a = A
                b = B
                c = np.min(y_fit)
                fit_success = True
            except Exception:
                fit_success = False

# Detect surge: if fitted b is significantly negative (price rises as days_left -> 0)

surge_detected = False
if fit_success and not np.isnan(b):

 # If b is negative and magnitude large enough, treat as surge (threshold chosen heuristically)

    surge_detected = (b < -0.01)

# Recommend booking window: choose days_left range with lowest median and low IQR
# Use median rolling window to smooth and find global minima region

agg['median'] = agg['price']
agg['median_smooth'] = agg['median'].rolling(window=7, min_periods=1, center=True).median()
min_idx = agg['median_smooth'].idxmin()
best_day = int(agg.loc[min_idx, 'days_left'])

# Recommend a small window around best_day (7 days), clipped to data range

recommended_window = (max(int(agg['days_left'].min()), best_day - 7),
                      min(int(agg['days_left'].max()), best_day + 7))


print("Fit success:", fit_success)
print("Fitted params (a, b, c):", (a, b, c))
print("Surge detected (heuristic):", surge_detected)
print("Recommended booking window (days left):", recommended_window)

# Example: show first few aggregated rows for context

print("\nSample median prices by days_left (first 15 rows):")
print(agg[['days_left', 'median']].head(15).to_string(index=False))

# If route comparison is needed, compute median price by airline for a route (Delhi -> Mumbai)

src_col = 'source' if 'source' in df.columns else ('origin' if 'origin' in df.columns else None)
dst_col = 'destination' if 'destination' in df.columns else ('dest' if 'dest' in df.columns else None)
if src_col and dst_col and 'airline' in df.columns:
    route_mask = (df[src_col].str.lower() == 'delhi') & (df[dst_col].str.lower() == 'mumbai')
    route_df = df[route_mask]
    if not route_df.empty:
        airline_stats = (route_df.groupby('airline', observed=True)['price']
                         .agg(median_price='median', q25=lambda x: x.quantile(0.25), q75=lambda x: x.quantile(0.75), count='count')
                         .sort_values('median_price'))
        print("\nDelhi -> Mumbai: median price by airline:")
        print(airline_stats.round(0).to_string())
    else:
        print("\nNo Delhi->Mumbai rows found in dataset for route comparison.")
else:
    print("\nRoute comparison skipped: required columns (source/destination and airline) not found.")
```

10. HR Analytics Dataset :-

a). What factors most strongly correlate with employee attrition? Use visualizations to show key drivers (e.g., satisfaction, overtime, salary).

b). Are employees with more projects more likely to leave?

Dataset: hr_analytics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# Load dataset

url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/e0da5fc18ebf4e3dc90950b932d6bded61e0ee49/hr_analytics.csv"
df = pd.read_csv(url)

# Basic cleaning and encoding

df = df.dropna().copy()
salary_map = {'low': 0, 'medium': 1, 'high': 2}
df['salary_code'] = df['salary'].map(salary_map)
df['sales'] = df['sales'].astype('category')

# Drop the original 'salary' column as its numerical representation 'salary_code' is now available

df = df.drop('salary', axis=1)

# Correlation with attrition
# Select only numeric columns for correlation calculation

numeric_df = df.select_dtypes(include=np.number)
corr_with_left = numeric_df.corr()['left'].sort_values(ascending=False)
print("Correlation with 'left':\n", corr_with_left)

# Visualizations

plt.figure(figsize=(10,6))

# Ensure only numeric columns relevant for heatmap are selected

heatmap_cols = ['satisfaction_level','last_evaluation','number_project',
                'average_montly_hours','time_spend_company','Work_accident',
                'promotion_last_5years','salary_code','left']
sns.heatmap(df[heatmap_cols].corr(),
            annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()

# Boxplots: satisfaction and salary by left

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(x='left', y='satisfaction_level', data=df)
plt.title('Satisfaction by Attrition')
plt.subplot(1,2,2)

# Note: for countplot, use the original 'salary' column if you want to display categories
# However, since we dropped 'salary', we'll use 'salary_code' for consistency in this corrected block
# If a categorical display is crucial here, 'salary_code' can be mapped back or a copy retained.

sns.countplot(x='salary_code', hue='left', data=df)
plt.title('Attrition counts by Salary')
plt.show()

# Attrition rate by overtime proxy: average_montly_hours > threshold

df['overtime'] = (df['average_montly_hours'] > df['average_montly_hours'].median()).astype(int)
overtime_rates = df.groupby('overtime')['left'].mean()
print("Attrition rate by overtime (1=high hours):\n", overtime_rates)

# Attrition rate by number of projects

proj_rates = df.groupby('number_project')['left'].agg(['count','mean']).reset_index()
print(proj_rates)
plt.figure(figsize=(8,4))
sns.barplot(x='number_project', y='mean', data=proj_rates)
plt.ylabel('Attrition rate')
plt.title('Attrition rate vs number of projects')
plt.show()

# Chi-square test for number_project vs left

cont = pd.crosstab(df['number_project'], df['left'])
chi2, p, dof, ex = chi2_contingency(cont)
print(f"Chi2 test: chi2={chi2:.2f}, p={p:.4f}")
```
